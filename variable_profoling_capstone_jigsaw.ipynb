```python
cat_variables = ['actvsubs', 'age1', 'age2', 'area', 'asl_flag', 'car_buy',
                 'crclscod', 'csa', 'ethnic', 'forgntvl', 'hnd_price', 'hnd_webcap',
                 'income', 'marital','models', 'mtrcycle', 'prizm_social_one',
                 'refurb_new', 'truck', 'uniqsubs', 'Customer_ID', 'churn']
```

```python
for variables in cat_variables:
    data[variables] = data[variables].fillna("Unknown")
    dataC1 = pd.DataFrame(data[variables].unique())
    dataC1['n'] = 0
    for i in range(len(dataC1)):
        dataC1.n[i] = len(data[data[variables]==dataC1[0][i]])
    dataC1['N'] = 0
    for i in range(len(dataC1)):
        dataC1.N[i] = len(data[(data[variables]==dataC1[0][i]) & (data["churn"]==1)])
    dataC1['pctc'] = dataC1.N/dataC1.n
    dataC1["variable"] = variables
    print("Profile of Variable:"+str(variables))
    print(dataC1)
```

```python
numeric_variables = ['mou_Mean', 'totmrc_Mean', 'rev_Range', 'mou_Range', 'change_mou',
       'drop_blk_Mean', 'drop_vce_Range', 'owylis_vce_Range', 'mou_opkv_Range',
       'months', 'totcalls', 'eqpdays', 'custcare_Mean',
       'callwait_Mean', 'iwylis_vce_Mean', 'callwait_Range', 'ccrndmou_Range',
       'adjqty', 'ovrrev_Mean', 'rev_Mean', 'ovrmou_Mean', 'comp_vce_Mean',
       'plcd_vce_Mean', 'avg3mou', 'avgmou', 'avg3qty', 'avgqty', 'avg6mou',
       'avg6qty','opk_dat_Mean', 'roam_Mean', 'recv_sms_Mean',
       'blck_dat_Mean', 'mou_pead_Mean', 'da_Mean',
       'da_Range', 'datovr_Mean', 'datovr_Range', 'drop_dat_Mean',
       'drop_vce_Mean', 'adjmou', 'totrev', 'adjrev', 'avgrev']
```

```python
for variables in numeric_variables:
    x = data[variables]
    data['variable_dec_bins'] = pd.qcut(x.rank(method='first'),10,labels=['A','B','C','D','E','F','G','H','I','J'])
    data['variable_dec_bins'] = data['variable_dec_bins'].astype('str')
    dataC1 = data['variable_dec_bins'].unique()
    dataC1 = pd.DataFrame(dataC1)
    dataC1['n'] = 0
    for i in range(len(dataC1)):
        dataC1.n[i] = len(data[data['variable_dec_bins']==dataC1[0][i]])
    dataC1['N'] = 0
    for i in range(len(dataC1)):
        dataC1.N[i] = len(data[(data['variable_dec_bins']==dataC1[0][i]) & (data["churn"]==1)])
    dataC1['pctc'] = round(dataC1.N/dataC1.n,2)
    dataC1["variable"] = variables
    dataC1['greaterthan']=0
    for i in range(len(dataC1)):
        dataC1['greaterthan'][i] = data[data['variable_dec_bins']==dataC1.iloc[i,0]][variables].min()
    dataC1['lessthan']=0
    for i in range(len(dataC1)):
        dataC1['lessthan'][i] = data[data['variable_dec_bins']==dataC1.iloc[i,0]][variables].max()
    dataC1.columns = ['decile','No. of Records','No. of Curn','Churn Pct','variable','greaterthan','lessthan']
    dataC1 = dataC1.sort_values(by = "lessthan")
    print("Profile of Variable:"+str(variables))
    print(dataC1.drop("variable",axis=1))
```
